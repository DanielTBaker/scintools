# Working with the EM simulations

### Developed from original Matlab code of Coles et al. (2010)


```python
# Import the Simulation class
from scintools.scint_sim import Simulation
```

First, let's just create an instance of the Simulation using default parameters, to plot and understand the results

```python
sim = Simulation()
```

```python
# Plot the simulated phase screen
sim.plot_screen()
```

```python
# Plot the observer-side intensity fluctuations in space (at the centre frequency)
sim.plot_intensity()
```

```python
# Plot the dynamic spectrum
sim.plot_dynspec()
```

```python
# Plot the dispersive group delay (at a fixed frequency) and the average impulse response function
sim.plot_delay()
```

```python
# Plot the (log) intensity of the scatter-broadened pulse in space
sim.plot_pulse()
```

Now, let's look at the input variables and then define a new Simulation object with our own parameters. Print the class description using the builtin help function

```python
help(sim.__init__)
```

Some key parameters we might want to change are below:

```python
mb2 = 20  # the Born variance, which controls the strength of scattering
ar = 2  # axial ratio of anisotropy
psi = 30  # angle of velocity relative to the x-axis

dlam = 0.33  # The fractional bandwidth
ns = 1024  # The size of the simulation in spatial steps. Total length in refractive scales is ns * ds
nf = 256  # The number of frequency channels across the dlam fractional bandwidth

seed = 1  # The seed to generate the random phase screen, for reproducible experiments
```

Let's compute and plot the resulting larger simulation

```python
my_sim = Simulation(mb2=mb2, ar=ar, psi=psi, dlam=dlam,
                    ns=ns, nf=nf, seed=seed)
```

```python
# Plot the phase screen and dynamic spectrum
my_sim.plot_screen()
my_sim.plot_dynspec()
```

```python
# Plot the delays and broadened pulse
my_sim.plot_delay()
my_sim.plot_pulse()
```

Here we can see a clear scattering tail because the strength of scattering was increased
